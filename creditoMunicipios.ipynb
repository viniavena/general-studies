```python
import requests
import pandas as pd
```

```python
url = "https://servicodados.ibge.gov.br/api/v1/localidades/estados/BA/municipios"

response = requests.request("GET", url)

municipios_response = response.text
```

```python
import json
import numpy as np

ano_de_exercicio = str(2019)
num_anexo = 'RREO-Anexo%2003'
num_municipio = str(2900108)

def GetItemsToDict(ano_de_exercicio, num_anexo, num_municipio):
  base_url = ["http://apidatalake.tesouro.gov.br/ords/siconfi/tt/rreo?","an_exercicio=",ano_de_exercicio,"&nr_periodo=6&co_tipo_demonstrativo=RREO&no_anexo",num_anexo,"&co_esfera=M&id_ente=",num_municipio]
  url = ''.join(base_url)
  response = requests.request("GET", url)
  response_dict = json.loads(response.text)
  return response_dict['items']
```

```python
nome_coluna = "TOTAL (ÚLTIMOS 12 MESES)"
cod_conta = 'ReceitaCorrenteLiquida'

def AchaReceita(dict_items, nome_coluna, cod_conta):
  receita = 0
  for item in dict_items:
    if (item['coluna'] == nome_coluna) and (cod_conta in item['cod_conta']):
      receita += item['valor']
      

  return receita

items = GetItemsToDict(ano_de_exercicio, num_anexo, num_municipio)
AchaReceita(items, nome_coluna, cod_conta)

municipios_array = municipio_df['id']

receitas = []

municipios_splitted = municipios_array[:5]

for id in municipios_splitted:
  id_str = str(id)
  items = GetItemsToDict(ano_de_exercicio, num_anexo, id_str)
  receita = AchaReceita(items, nome_coluna, cod_conta)
  receitas.append(round(receita,2))
  print('id: ', id, 'receita: ',receita)



```
